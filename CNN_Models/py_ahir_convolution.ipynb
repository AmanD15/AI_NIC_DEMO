```python
def convolution(groups,in_channels,reuseInp):
    convOutVal = 0
    for grp_no in range(groups):
        __loop_pipelining_on__(31, 2, 1)

        # -------------------------------------Slice 1 Start----------------------------------------------- #
        # Convolution of Cylinder (0,0)
        # print(grp_no)
        inp_val_0_0 = [channel.read_float32("in_data_0") for _ in range(8)]
        # print(inp_val_0_0)
        ker_val_0_0 = [channel.read_float32("ker_data") for _ in range(8)]
        [channel.write_float32("ker_data",val) for val in ker_val_0_0]

        mul_val_0 = inp_val_0_0[0] * ker_val_0_0[0]
        mul_val_1 = inp_val_0_0[1] * ker_val_0_0[1]
        mul_val_2 = inp_val_0_0[2] * ker_val_0_0[2]
        mul_val_3 = inp_val_0_0[3] * ker_val_0_0[3]
        mul_val_4 = inp_val_0_0[4] * ker_val_0_0[4]
        mul_val_5 = inp_val_0_0[5] * ker_val_0_0[5]
        mul_val_6 = inp_val_0_0[6] * ker_val_0_0[6]
        mul_val_7 = inp_val_0_0[7] * ker_val_0_0[7]

        acc_val_0_0 = ( ((mul_val_0 + mul_val_1) +(mul_val_2 + mul_val_3)) +
                        ((mul_val_4 + mul_val_5) +(mul_val_6 + mul_val_7)) )

        # Convolution of Cylinder (1,0)
        inp_val_1_0 = [channel.read_float32("in_data_0") for _ in range(8)]
        ker_val_1_0 = [channel.read_float32("ker_data") for _ in range(8)]
        [channel.write_float32("ker_data",val) for val in ker_val_1_0]

        mul_val_0 = inp_val_1_0[0] * ker_val_1_0[0]
        mul_val_1 = inp_val_1_0[1] * ker_val_1_0[1]
        mul_val_2 = inp_val_1_0[2] * ker_val_1_0[2]
        mul_val_3 = inp_val_1_0[3] * ker_val_1_0[3]
        mul_val_4 = inp_val_1_0[4] * ker_val_1_0[4]
        mul_val_5 = inp_val_1_0[5] * ker_val_1_0[5]
        mul_val_6 = inp_val_1_0[6] * ker_val_1_0[6]
        mul_val_7 = inp_val_1_0[7] * ker_val_1_0[7]

        acc_val_1_0 = ( ((mul_val_0 + mul_val_1) +(mul_val_2 + mul_val_3)) +
                        ((mul_val_4 + mul_val_5) +(mul_val_6 + mul_val_7)) )

        # Convolution of Cylinder (2,0)
        inp_val_2_0 = [channel.read_float32("in_data_0") for _ in range(8)]
        ker_val_2_0 = [channel.read_float32("ker_data") for _ in range(8)]
        [channel.write_float32("ker_data",val) for val in ker_val_2_0]


        mul_val_0 = inp_val_2_0[0] * ker_val_2_0[0]
        mul_val_1 = inp_val_2_0[1] * ker_val_2_0[1]
        mul_val_2 = inp_val_2_0[2] * ker_val_2_0[2]
        mul_val_3 = inp_val_2_0[3] * ker_val_2_0[3]
        mul_val_4 = inp_val_2_0[4] * ker_val_2_0[4]
        mul_val_5 = inp_val_2_0[5] * ker_val_2_0[5]
        mul_val_6 = inp_val_2_0[6] * ker_val_2_0[6]
        mul_val_7 = inp_val_2_0[7] * ker_val_2_0[7]

        acc_val_2_0 = ( ((mul_val_0 + mul_val_1) +(mul_val_2 + mul_val_3)) +
                        ((mul_val_4 + mul_val_5) +(mul_val_6 + mul_val_7)) )
        # -------------------------------------Slice 1 End----------------------------------------------- #

        # -------------------------------------Slice 2 Start----------------------------------------------- #
        # Convolution of Cylinder (0,1)
        inp_val_0_1 = [channel.read_float32("in_data_1") for _ in range(8)]
        # print(inp_val_0_1)
        ker_val_0_1 = [channel.read_float32("ker_data") for _ in range(8)]
        [channel.write_float32("ker_data",val) for val in ker_val_0_1]

        mul_val_0 = inp_val_0_1[0] * ker_val_0_1[0]
        mul_val_1 = inp_val_0_1[1] * ker_val_0_1[1]
        mul_val_2 = inp_val_0_1[2] * ker_val_0_1[2]
        mul_val_3 = inp_val_0_1[3] * ker_val_0_1[3]
        mul_val_4 = inp_val_0_1[4] * ker_val_0_1[4]
        mul_val_5 = inp_val_0_1[5] * ker_val_0_1[5]
        mul_val_6 = inp_val_0_1[6] * ker_val_0_1[6]
        mul_val_7 = inp_val_0_1[7] * ker_val_0_1[7]

        acc_val_0_1 = ( ((mul_val_0 + mul_val_1) +(mul_val_2 + mul_val_3)) +
                        ((mul_val_4 + mul_val_5) +(mul_val_6 + mul_val_7)) )

        # Convolution of Cylinder (1,1)
        inp_val_1_1 = [channel.read_float32("in_data_1") for _ in range(8)]
        ker_val_1_1 = [channel.read_float32("ker_data") for _ in range(8)]
        [channel.write_float32("ker_data",val) for val in ker_val_1_1]

        mul_val_0 = inp_val_1_1[0] * ker_val_1_1[0]
        mul_val_1 = inp_val_1_1[1] * ker_val_1_1[1]
        mul_val_2 = inp_val_1_1[2] * ker_val_1_1[2]
        mul_val_3 = inp_val_1_1[3] * ker_val_1_1[3]
        mul_val_4 = inp_val_1_1[4] * ker_val_1_1[4]
        mul_val_5 = inp_val_1_1[5] * ker_val_1_1[5]
        mul_val_6 = inp_val_1_1[6] * ker_val_1_1[6]
        mul_val_7 = inp_val_1_1[7] * ker_val_1_1[7]

        acc_val_1_1 = ( ((mul_val_0 + mul_val_1) +(mul_val_2 + mul_val_3)) +
                        ((mul_val_4 + mul_val_5) +(mul_val_6 + mul_val_7)) )

        # Convolution of Cylinder (2,1)
        inp_val_2_1 = [channel.read_float32("in_data_1") for _ in range(8)]
        ker_val_2_1 = [channel.read_float32("ker_data") for _ in range(8)]
        [channel.write_float32("ker_data",val) for val in ker_val_2_1]

        mul_val_0 = inp_val_2_1[0] * ker_val_2_1[0]
        mul_val_1 = inp_val_2_1[1] * ker_val_2_1[1]
        mul_val_2 = inp_val_2_1[2] * ker_val_2_1[2]
        mul_val_3 = inp_val_2_1[3] * ker_val_2_1[3]
        mul_val_4 = inp_val_2_1[4] * ker_val_2_1[4]
        mul_val_5 = inp_val_2_1[5] * ker_val_2_1[5]
        mul_val_6 = inp_val_2_1[6] * ker_val_2_1[6]
        mul_val_7 = inp_val_2_1[7] * ker_val_2_1[7]

        acc_val_2_1 = ( ((mul_val_0 + mul_val_1) +(mul_val_2 + mul_val_3)) +
                        ((mul_val_4 + mul_val_5) +(mul_val_6 + mul_val_7)) )
        # -------------------------------------Slice 2 End----------------------------------------------- #

        # -------------------------------------Slice 3 Start----------------------------------------------- #
        # Convolution of Cylinder (0,2)
        inp_val_0_2 = [channel.read_float32("in_data_2") for _ in range(8)]
        ker_val_0_2 = [channel.read_float32("ker_data") for _ in range(8)]
        [channel.write_float32("ker_data",val) for val in ker_val_0_2]

        mul_val_0 = inp_val_0_2[0] * ker_val_0_2[0]
        mul_val_1 = inp_val_0_2[1] * ker_val_0_2[1]
        mul_val_2 = inp_val_0_2[2] * ker_val_0_2[2]
        mul_val_3 = inp_val_0_2[3] * ker_val_0_2[3]
        mul_val_4 = inp_val_0_2[4] * ker_val_0_2[4]
        mul_val_5 = inp_val_0_2[5] * ker_val_0_2[5]
        mul_val_6 = inp_val_0_2[6] * ker_val_0_2[6]
        mul_val_7 = inp_val_0_2[7] * ker_val_0_2[7]

        acc_val_0_2 = ( ((mul_val_0 + mul_val_1) +(mul_val_2 + mul_val_3)) +
                        ((mul_val_4 + mul_val_5) +(mul_val_6 + mul_val_7)) )

        # Convolution of Cylinder (1,1)
        inp_val_1_2 = [channel.read_float32("in_data_2") for _ in range(8)]
        ker_val_1_2 = [channel.read_float32("ker_data") for _ in range(8)]
        [channel.write_float32("ker_data",val) for val in ker_val_1_2]

        mul_val_0 = inp_val_1_2[0] * ker_val_1_2[0]
        mul_val_1 = inp_val_1_2[1] * ker_val_1_2[1]
        mul_val_2 = inp_val_1_2[2] * ker_val_1_2[2]
        mul_val_3 = inp_val_1_2[3] * ker_val_1_2[3]
        mul_val_4 = inp_val_1_2[4] * ker_val_1_2[4]
        mul_val_5 = inp_val_1_2[5] * ker_val_1_2[5]
        mul_val_6 = inp_val_1_2[6] * ker_val_1_2[6]
        mul_val_7 = inp_val_1_2[7] * ker_val_1_2[7]

        acc_val_1_2 = ( ((mul_val_0 + mul_val_1) +(mul_val_2 + mul_val_3)) +
                        ((mul_val_4 + mul_val_5) +(mul_val_6 + mul_val_7)) )

        # Convolution of Cylinder (2,1)
        inp_val_2_2 = [channel.read_float32("in_data_2") for _ in range(8)]
        ker_val_2_2 = [channel.read_float32("ker_data") for _ in range(8)]
        [channel.write_float32("ker_data",val) for val in ker_val_2_2]

        mul_val_0 = inp_val_2_2[0] * ker_val_2_2[0]
        mul_val_1 = inp_val_2_2[1] * ker_val_2_2[1]
        mul_val_2 = inp_val_2_2[2] * ker_val_2_2[2]
        mul_val_3 = inp_val_2_2[3] * ker_val_2_2[3]
        mul_val_4 = inp_val_2_2[4] * ker_val_2_2[4]
        mul_val_5 = inp_val_2_2[5] * ker_val_2_2[5]
        mul_val_6 = inp_val_2_2[6] * ker_val_2_2[6]
        mul_val_7 = inp_val_2_2[7] * ker_val_2_2[7]

        acc_val_2_2 = ( ((mul_val_0 + mul_val_1) +(mul_val_2 + mul_val_3)) +
                        ((mul_val_4 + mul_val_5) +(mul_val_6 + mul_val_7)) )
        # -------------------------------------Slice 3 End----------------------------------------------- #
        convOutVal += (acc_val_0_0 + acc_val_1_0 + acc_val_2_0) + (acc_val_0_1 + acc_val_1_1 + acc_val_2_1) +(acc_val_0_2 + acc_val_1_2 + acc_val_2_2)
        if reuseInp == 1: # Forward Movement
          [channel.write_float32("in_data_0",val) for val in inp_val_0_1]
          [channel.write_float32("in_data_0",val) for val in inp_val_1_1]
          [channel.write_float32("in_data_0",val) for val in inp_val_2_1]


          [channel.write_float32("in_data_1",val) for val in inp_val_0_2]
          [channel.write_float32("in_data_1",val) for val in inp_val_1_2]
          [channel.write_float32("in_data_1",val) for val in inp_val_2_2]

        elif reuseInp == 2: # Reverse Movement
          [channel.write_float32("in_data_1",val) for val in inp_val_0_0]
          [channel.write_float32("in_data_1",val) for val in inp_val_1_0]
          [channel.write_float32("in_data_1",val) for val in inp_val_2_0]

          [channel.write_float32("in_data_2",val) for val in inp_val_0_1]
          [channel.write_float32("in_data_2",val) for val in inp_val_1_1]
          [channel.write_float32("in_data_2",val) for val in inp_val_2_1]
        elif reuseInp == 0: # Cycle the input data
          # [channel.write_float32("in_data_0",val) for val in inp_val_0_0]
          [channel.write_float32("in_data_0",val) for val in inp_val_1_0]
          [channel.write_float32("in_data_0",val) for val in inp_val_2_0]

          # [channel.write_float32("in_data_1",val) for val in inp_val_0_1]
          [channel.write_float32("in_data_1",val) for val in inp_val_1_1]
          [channel.write_float32("in_data_1",val) for val in inp_val_2_1]

          # [channel.write_float32("in_data_2",val) for val in inp_val_0_2]
          [channel.write_float32("in_data_2",val) for val in inp_val_1_2]
          [channel.write_float32("in_data_2",val) for val in inp_val_2_2]





    return convOutVal
```

```python
def fetchKernel(groups, ker_size, in_channels):
  grp_no = 0
  while grp_no < groups:
    for col in range(ker_size):
        for row in range(ker_size):
          for chn in range(8*grp_no,8*(grp_no+1)):
            if(chn>=in_channels):
              channel.write_float32("ker_data",0)
              continue
            channel.write_float32("ker_data",ker[chn][row][col])
    grp_no+=1
```

```python
def fetchInpGroup(startRowIndex,startColIndex,grp_no,ker_size,padReq,in_rows,in_cols,in_channels):
    for col in range(ker_size):
      if col  == 0:
        pipeName = "in_data_0"
        # print(pipeName)
      elif col == 1:
        pipeName = "in_data_1"
      elif col == 2:
        pipeName = "in_data_2"
      for row in range(ker_size):
          for chn in range(8*grp_no,8*(grp_no+1)):
              if(chn>=in_channels):
                  channel.write_float32(pipeName,0)
                  continue
              # channel.write_float32(inPipeName,inp[chn][startRowIndex+row][startColIndex+col])
              #for padding, whenever we are on 0th or last row/col, we return a 0 cylinder
              if padReq:
                if((startRowIndex+row)==0 or (startColIndex+col)==0 or
                    (startRowIndex+row) == (in_rows+1) or (startColIndex+col)==(in_cols+1)):
                    channel.write_float32(pipeName,0)
                else:
                    channel.write_float32(pipeName,inp[chn][startRowIndex+row-1][startColIndex+col-1])
              else:
                channel.write_float32(pipeName,inp[chn][startRowIndex+row][startColIndex+col])
                # if col ==0:
                #   print(pipeName)

    return
```

```python
def fetchInpVerSlice(startRowIndex,startColIndex,grp_no,ker_size,padReq,in_rows,in_cols,in_channels):
  col = 0
  # print("At Start ",startColIndex)
  if (startRowIndex == 0) and (startColIndex  == 0):
    pipeName = "in_data_0"
    # print(pipeName)
  elif (startRowIndex == 0) and (startColIndex == 1):
    pipeName = "in_data_1"
  else:
    if startRowIndex%2 == 0: # Forward Movement
      pipeName = "in_data_2"
    else: # Reverse Movement
      pipeName = "in_data_0"
  # print(pipeName)
  for row in range(ker_size):
      for chn in range(8*grp_no,8*(grp_no+1)):
          if(chn>=in_channels):
              channel.write_float32(pipeName,0)
              continue
          # channel.write_float32(inPipeName,inp[chn][startRowIndex+row][startColIndex+col])
          #for padding, whenever we are on 0th or last row/col, we return a 0 cylinder
          if padReq:
            if((startRowIndex+row)==0 or (startColIndex+col)==0 or
                (startRowIndex+row) == (in_rows+1) or (startColIndex+col)==(in_cols+1)):
              channel.write_float32(pipeName,0)
            else:
              # print("Infunc end",startColIndex)
              channel.write_float32(pipeName,inp[chn][startRowIndex+row-1][startColIndex+col-1])
          else:
            # print(pipeName)
            channel.write_float32(pipeName,inp[chn][startRowIndex+row][startColIndex+col])
              # if col ==0:
              #   print(pipeName)
  return
```

```python
def fetchInpHorSlice(startRowIndex,grp_no,ker_size,padReq,in_rows,in_cols,in_channels,out_cols):
  row = ker_size

  if startRowIndex%2 == 0:
    startColIndex = out_cols-1
  else:
    startColIndex = 0

  for col in range(ker_size):
    if col  == 0:
      pipeName = "in_data_0"
    elif col == 1:
      pipeName = "in_data_1"
    elif col == 2:
      pipeName = "in_data_2"

    r2_3_vals = [channel.read_float32(pipeName) for _ in range(16)]
    [channel.write_float32(pipeName,val) for val in r2_3_vals]
    for chn in range(8*grp_no,8*(grp_no+1)):
        if(chn>=in_channels):
            channel.write_float32(pipeName,0)
            continue
        # channel.write_float32(inPipeName,inp[chn][startRowIndex+row][startColIndex+col])
        #for padding, whenever we are on 0th or last row/col, we return a 0 cylinder
        if padReq:
          if((startRowIndex+row)==0 or (startColIndex+col)==0 or
              (startRowIndex+row) == (in_rows+1) or (startColIndex+col)==(in_cols+1)):
            channel.write_float32(pipeName,0)
          else:
            # print("Infunc end",startColIndex)
            channel.write_float32(pipeName,inp[chn][startRowIndex+row-1][startColIndex+col-1])
        else:

          channel.write_float32(pipeName,inp[chn][startRowIndex+row][startColIndex+col])
  return
```

```python
def pooling(poolFirstStage,row_no):
  global max_val_col0,max_val_col1,max_val_col2
  maxPoolVal = 0
  if poolFirstStage:
    r0_fifo_vals = [channel.read_float32("pool_data_fifo_r0") for _ in range(3)]

    r1_lifo_vals = [channel.read_float32("pool_data_lifo_r1") for _ in range(3)]
    r1_fifo_vals = [channel.read_float32("pool_data_fifo_r1") for _ in range(3)]
    channel.write_float32("pool_data_fifo_r0",r1_fifo_vals[0])
    channel.write_float32("pool_data_fifo_r0",r1_fifo_vals[1])
    channel.write_float32("pool_data_fifo_r0",r1_fifo_vals[2])

    r2_fifo_vals = [channel.read_float32("pool_data_fifo_r2") for _ in range(3)]
    channel.write_float32("pool_data_fifo_r1",r2_fifo_vals[0])
    channel.write_float32("pool_data_fifo_r1",r2_fifo_vals[1])
    channel.write_float32("pool_data_fifo_r1",r2_fifo_vals[2])

    max_val_col0 = max(r0_fifo_vals[0],r1_lifo_vals[0],r2_fifo_vals[0])
    max_val_col1 = max(r0_fifo_vals[1],r1_lifo_vals[1],r2_fifo_vals[1])
    max_val_col2 = max(r0_fifo_vals[2],r1_lifo_vals[2],r2_fifo_vals[2])

    maxPoolVal = max(max_val_col0,max_val_col1,max_val_col2)

    max_val_col0 = max_val_col1
    max_val_col1 = max_val_col2

  else:
    r0_fifo_val = channel.read_float32("pool_data_fifo_r0")

    r1_lifo_val = channel.read_float32("pool_data_lifo_r1")
    r1_fifo_val = channel.read_float32("pool_data_fifo_r1")
    channel.write_float32("pool_data_fifo_r0",r1_fifo_val)

    r2_fifo_val = channel.read_float32("pool_data_fifo_r2")
    channel.write_float32("pool_data_fifo_r1",r2_fifo_val)
    # channel.write_float32("pool_data_lifo_r1",r2_fifo_val)

    max_val_col2 = max(r0_fifo_val,r1_lifo_val,r2_fifo_val)


    maxPoolVal = max(max_val_col0,max_val_col1,max_val_col2)

    max_val_col0 = max_val_col1
    max_val_col1 = max_val_col2

  return maxPoolVal

```

```python
def conv_engine(in_channels,out_rows,out_cols,groups):
    convOutVal = 0
    for row_no in range(out_rows):
        for col_no in range(out_cols):
            convOutVal = convolution(groups,in_channels)
            print("Convolution Done and value is ", convOutVal)
            channel.write_float32("out_data",convOutVal)
    return
```
